```python
cbct_input_dir= "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/cbct_input"
ct_gt_dir = "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/ct_gt"

test_dirs = [
    "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/net_ct_out",
    "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/rescale0to1_net_out_epoch50000",
    "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/default_loss_epoch50000",
    "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/l1_gan_loss_epoch40000",
    "/home/liusheng/worksapce2024/github/Real-ESRGAN/datasets/CBCT/test_set/only_l1_loss_epoch50000",
]

filenames = natsorted(os.listdir(cbct_input_dir))

idx = 60
file = filenames[idx]
# file = "2PC094_20.npy"
print(f"filename: {file}")

cbct_img_path = os.path.join(cbct_input_dir, file)
ct_gt_path = os.path.join(ct_gt_dir, file)

cbct_img = np.load(cbct_img_path)[:, ::-1]
ct_gt_img = np.load(ct_gt_path)[:, ::-1]
```

```python
plt.figure(figsize=(20, 10))

vmin = None
vmax = None

cols = len(test_dirs) + 1

ax = plt.subplot(2, cols, 1)
im = ax.imshow(cbct_img, cmap='gray', vmin=vmin, vmax=vmax)
plt.colorbar(im, ax=ax, shrink=0.6)
ax.set_title("CBCT Image")

ax = plt.subplot(2, cols, cols + 1)
im = ax.imshow(ct_gt_img, cmap='gray', vmin=vmin, vmax=vmax)
plt.colorbar(im, ax=ax, shrink=0.6)
ax.set_title("CT GT Image")

for idx, test_dir in enumerate(test_dirs):
    head, title = os.path.split(test_dir)
    test_path = os.path.join(test_dir, file)
    test_img = np.load(test_path)[:, ::-1, 0]
    
    ax = plt.subplot(2, cols, idx + 2)
    im = ax.imshow(test_img, cmap='gray', vmin=vmin, vmax=vmax)
    plt.colorbar(im, ax=ax, shrink=0.6)
    ax.set_title(title)
    
    ax = plt.subplot(2, cols, idx + 2 + cols)
    im = ax.imshow(test_img - ct_gt_img, cmap='gray')
    plt.colorbar(im, ax=ax, shrink=0.6)
    ax.set_title("Pred Image - CT GT")

plt.suptitle(file)
plt.tight_layout()
```
